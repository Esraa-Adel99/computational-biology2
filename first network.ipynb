```python
import networkx as nx
```

```python
g =nx.Graph()
```

```python
g.add_nodes_from(['a','b','c','d'])
```

```python
g.add_edges_from([('a','b'),('a','c'),('c','d'),('b','d'),('a','d'),('b','c')])
```

```python
nx.draw(g,with_labels=True,node_color='red',node_size=2000,font_size=20,font_color='white')
```
